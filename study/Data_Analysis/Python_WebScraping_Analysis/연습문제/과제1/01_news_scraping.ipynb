```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/economy'

req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=req_header)
res.encoding = 'utf-8'

print(url)
print(type(res))
print(res.status_code)

if res.ok:
    soup = BeautifulSoup(res.text, "html.parser")

    li_tags = soup.select('ul.list_newsheadline2 li')
    print(type(li_tags), len(li_tags))

    for li_tag in li_tags:
        a_tag = li_tag.find('a')
        if not a_tag:
            continue
        link = a_tag['href']

        strong_tag = li_tag.select_one('div.wrap_thumb strong.tit_txt')
        if not strong_tag:
            title = a_tag.get_text(strip=True)
        else:
            title = strong_tag.text.strip()

        print(link)
        print(title)

else:
    print(f"Error Code = {res.status_code}")
```

```python
'''print_news() 함수에서 
   section name (섹션명)을 입력 받아서 section_dict에서 section 영문   
   문자열을 추출해서 url을 생성합니다.
   그 이후의 코드는 질문1번 문제와 동일합니다.
'''

section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics',\
             '국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}


def print_news(section_name):
    sid = section_dict[section_name]
    url = f'https://news.daum.net/{sid}'
    print(f"======> {url} {section_name} 뉴스 <======")

    req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if res.ok:
        soup = BeautifulSoup(res.text, "html.parser")

        li_tags = soup.select('ul.list_newsheadline2 li')

        for li_tag in li_tags:
            a_tag = li_tag.find('a')
            if not a_tag:
                continue
            link = a_tag['href']

            strong_tag = li_tag.select_one('div.wrap_thumb strong.tit_txt')
            if not strong_tag:
                title = a_tag.get_text(strip=True)
            else:
                title = strong_tag.text.strip()

            print(link)
            print(title)

    else:
        print(f"Error Code = {res.status_code}")
```

```python
# 함수호출
print_news('경제')
print()
print_news('사회')
```
